- home_team_history_match_data_{i} - Дата последнего восьмого матча, сыгранного домашней командой.
- home_team_history_is_play_home_{i} - Если 1, то команда хозяев играла дома.
- home_team_history_is_cup_{i} - Если 1, то матч был кубковым соревнованием.
- home_team_history_goal_{i} - количество голов, забитых командой хозяев поля в последних восьми матчах.
- home_team_history_opponent_goal_{i} - количество голов, пропущенных хозяевами поля в последних восьми матчах.
- home_team_history_rating_{i} - рейтинг домашней команды в последних восьми матчах (предматчевый рейтинг).
- home_team_history_opponent_rating_{i} - Рейтинг команды соперника, в которой хозяева проиграли матч (предматчевый рейтинг).
- home_team_history_coach_{i} - Идентификатор тренера домашней команды в ее последних восьми матчах.
- home_team_history_league_id_{i} - Идентификатор названия лиги, присвоенный домашней командой в ее последних восьми матчах.
